### **–ó–∞–¥–∞—á–∞ 4. –§—É–Ω–∫—Ü–∏—è**

### **–ß—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å**
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª—É—á–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –Ω–∞—á–∞–ª–æ –∏ –∫–æ–Ω–µ—Ü –æ—Ç—Ä–µ–∑–∫–∞, –∞ —Ç–∞–∫–∂–µ —à–∞–≥ (–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π), –∞ –∑–∞—Ç–µ–º –≤—ã—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é y –≤ –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–µ –æ—Ç—Ä–µ–∑–∫–∞ –∏ –≤—ã–≤–æ–¥–∏—Ç –æ—Ç–≤–µ—Ç –Ω–∞ —ç–∫—Ä–∞–Ω —Å –Ω—É–∂–Ω—ã–º —à–∞–≥–æ–º.

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ,** —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –º–æ–∂–µ—Ç –≤–≤–µ—Å—Ç–∏ –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∏–ª–∏ —à–∞–≥–∞. –í–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø—Ä–∞–≤–∏—Ç—å —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ–±—ã –≤—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –Ω–µ –∑–∞–≤–∏—Å–µ–ª –æ—Ç –æ—à–∏–±–æ—á–Ω–æ –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

–°–∞–º–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≥–ª—è–¥–∏—Ç —Ç–∞–∫: `y = ùë•¬≥ + 2 ‚àô ùë•¬≤ - 4 ‚àô x + 1 `

–í —Ç–æ—á–∫–µ ‚àí2 —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞–≤–Ω–∞ 9

**‚Äå–ü—Ä–∏–º–µ—Ä 2 (–ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤—ë–ª –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —à–∞–≥)**

–í–≤–µ–¥–∏—Ç–µ –∫–æ–Ω–µ—Ü –æ—Ç—Ä–µ–∑–∫–∞: 9

#### **–ß—Ç–æ –æ—Ü–µ–Ω–∏–≤–∞–µ—Ç—Å—è**
- –†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã–≤–æ–¥–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —É—Å–ª–æ–≤–∏—é.
- –í—ã–≤–æ–¥ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–ø–∏—Å–∞–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ (–≤—ã–≤–µ–¥–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ —Å–æ–ø—Ä–æ–≤–æ–∂–¥–∞—é—Ç—Å—è —Ç–µ–∫—Å—Ç–æ–≤—ã–º –æ–ø–∏—Å–∞–Ω–∏–µ–º).
- –í –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø—Ä–æ–≤–µ—Ä–∫–∞ –≥—Ä–∞–Ω–∏—Ü –ø–µ—Ä–µ–¥ —Å–æ–∑–¥–∞–Ω–∏–µ–º range –∏, –≤–æ–∑–º–æ–∂–Ω–æ, –∏—Ö —Ä–∞–∑–≤–æ—Ä–æ—Ç, –∫–∞–∫ –∏ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –∑–Ω–∞–∫–∞ —à–∞–≥–∞.
- –§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ø—Ä–∏–º–µ—Ä—É.




````python
# –ó–¥–µ—Å—å –Ω–∞–ø–∏—à–∏—Ç–µ –∫–æ–¥ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ 4
def calculate_function(x):
    return x**3 + 2 * x**2 - 4 * x + 1


def main():
    try:
        # –í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        start = float(input("–í–≤–µ–¥–∏—Ç–µ –Ω–∞—á–∞–ª–æ –æ—Ç—Ä–µ–∑–∫–∞: "))
        end = float(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–Ω–µ—Ü –æ—Ç—Ä–µ–∑–∫–∞: "))
        step = float(input("–í–≤–µ–¥–∏—Ç–µ —à–∞–≥ (–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π): "))

        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
        if step >= 0:
            print("–®–∞–≥ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º. –ò—Å–ø—Ä–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ —à–∞–≥–∞...")
            step = -abs(step)

        if start < end:
            print("–ù–∞—á–∞–ª–æ –æ—Ç—Ä–µ–∑–∫–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –º–µ–Ω—å—à–µ –∫–æ–Ω—Ü–∞ –ø—Ä–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–º —à–∞–≥–µ. –ú–µ–Ω—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è...")
            start, end = end, start

        # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –∑–∞–¥–∞–Ω–Ω–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ
        print("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏–π:")
        x = start
        while x >= end:
            y = calculate_function(x)
            print(f"–í —Ç–æ—á–∫–µ = {x:.0f}, —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞–≤–Ω–∞ = {y:.0f}")
            x += step

    except ValueError:
        print("–û—à–∏–±–∫–∞: –≤–≤–µ–¥–µ–Ω—ã –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –≤–≤–æ–¥–∏—Ç–µ —Ç–æ–ª—å–∫–æ —á–∏—Å–ª–∞.")


# –ó–∞–ø—É—Å–∫ –ø—Ä–æ–≥—Ä–∞–º–º—ã
if __name__ == "__main__":
    main()
````

- –†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã–≤–æ–¥–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —É—Å–ª–æ–≤–∏—é.
- –í—ã–≤–æ–¥ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–ø–∏—Å–∞–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ (–≤—ã–≤–µ–¥–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ —Å–æ–ø—Ä–æ–≤–æ–∂–¥–∞—é—Ç—Å—è —Ç–µ–∫—Å—Ç–æ–≤—ã–º –æ–ø–∏—Å–∞–Ω–∏–µ–º).


````python
# –ó–¥–µ—Å—å –Ω–∞–ø–∏—à–∏—Ç–µ –∫–æ–¥ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ 5
# –í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
educational_grant = int(input("–í–≤–µ–¥–∏—Ç–µ –µ–∂–µ–º–µ—Å—è—á–Ω—É—é —Å—Ç–∏–ø–µ–Ω–¥–∏—é: "))
expenses = int(input("–í–≤–µ–¥–∏—Ç–µ –µ–∂–µ–º–µ—Å—è—á–Ω—ã–µ —Ä–∞—Å—Ö–æ–¥—ã: "))

total_deficit = 0  # –û–±—â–∞—è —Å—É–º–º–∞, –∫–æ—Ç–æ—Ä—É—é –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —É —Ä–æ–¥–∏—Ç–µ–ª–µ–π

# –†–∞—Å—á—ë—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–µ—Å—è—Ü–∞
for month in range(1, 11):
    if month == 1:
        current_expenses = expenses  # –í –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü —Ä–∞—Å—Ö–æ–¥—ã –Ω–µ —É–≤–µ–ª–∏—á–∏–≤–∞—é—Ç—Å—è
    else:
        current_expenses = int(current_expenses * 1.03)  # –£–≤–µ–ª–∏—á–µ–Ω–∏–µ —Ä–∞—Å—Ö–æ–¥–æ–≤ –Ω–∞ 3%

    deficit = current_expenses - educational_grant  # –ù–µ–¥–æ—Å—Ç–∞—é—â–∞—è —Å—É–º–º–∞
    total_deficit += deficit  # –î–æ–±–∞–≤–ª—è–µ–º –∫ –æ–±—â–µ–π —Å—É–º–º–µ

    print(f"{month}-–π –º–µ—Å—è—Ü: —Ç—Ä–∞—Ç—ã {current_expenses} —Ä—É–±–ª–µ–π, –Ω–µ —Ö–≤–∞—Ç–∞–µ—Ç {deficit} —Ä—É–±–ª–µ–π.")

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print(f"–°—É–º–º–∞ –¥–µ–Ω–µ–≥, –∫–æ—Ç–æ—Ä—É—é –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–ª—É—á–∏—Ç—å —É —Ä–æ–¥–∏—Ç–µ–ª–µ–π: {total_deficit} —Ä—É–±–ª–µ–π")
````
